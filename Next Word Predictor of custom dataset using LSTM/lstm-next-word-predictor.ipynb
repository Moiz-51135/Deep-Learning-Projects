```python
input_sequences = []
for sentence in data.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max([len(x) for x in input_sequences])
```

```python
max_len = max([len(x) for x in input_sequences])
```

# Padding upto total of 57 digits because max words in a Sentence=57


```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
sequence_lengths = [len(seq) for seq in input_sequences]
plt.figure(figsize=(8, 6))
plt.hist(sequence_lengths, bins=30, color='green', alpha=0.7)
plt.title('Sequence Length Distribution')
plt.xlabel('Sequence Length')
plt.ylabel('Frequency')
plt.show()
```

```python
padded_input_sequences
```

# Converting in a Supervised Learning Problem so that input gives output

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

# One Hot Encoding

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=310)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(310, 100, input_length=37))
model.add(LSTM(150))
model.add(Dense(310, activation='softmax'))
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(X, y, epochs=100)
plt.figure(figsize=(12, 6))
plt.plot(history.history['accuracy'], label='Accuracy')
plt.plot(history.history['loss'], label='Loss')
plt.title('Model Training History')
plt.xlabel('Epochs')
plt.legend()
plt.show()
```

```python
import numpy as np
import time
import matplotlib.pyplot as plt

text = "Where can we contact you?"
predicted_words = []

for i in range(15):
    token_text = tokenizer.texts_to_sequences([text])[0]
    padded_token_text = pad_sequences([token_text], maxlen=37, padding='pre')
    pos = np.argmax(model.predict(padded_token_text))

    for word, index in tokenizer.word_index.items():
        if index == pos:
            text = text + " " + word
            predicted_words.append(word)
    
    # Plotting the progress
    plt.figure(figsize=(12, 6))
    plt.plot(range(len(predicted_words)), predicted_words, marker='o', linestyle='-', color='b')
    plt.title('Word Prediction Progress')
    plt.xlabel('Iterations')
    plt.ylabel('Predicted Words')
    plt.grid(True)
    plt.show()
    
    time.sleep(2)
```
